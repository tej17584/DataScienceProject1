# Carga de Librerías

```python
import numpy as np # linear algebra
import pandas as pd # para analisis de CSV
import tensorflow as tf #tensor flow para imgenes
import datetime, os #datetime
import math #para calculos
import matplotlib.pyplot as plt #para graficar
import seaborn as sns #para lindas graficas
from sklearn.model_selection import train_test_split
```

# Carga de data

```python
#Cargamos los dataframes
train_df = pd.read_csv('/kaggle/input/rsna-bone-age/boneage-training-dataset.csv')
test_df = pd.read_csv('/kaggle/input/rsna-bone-age/boneage-test-dataset.csv')
print("Done")
```

Miramos los valores del train y test superficialmente. 

```python
train_df.head()
```

Notamos que tenemos 3 variables en total:
* id: es el id de la imagen, su identificador unico.
* boneage: es el calculo de la edad osea del paciente.
* male: true or false. Si es true, es hombre, si es false es mujer.

Aplicaremos un append al grupo de la variable id. Le agregaremos el sufijo: .png. Esto con la finalidad de que al momento de mapear el valor con las imágenes, estas coincidan.

```python
train_df['id'] = train_df['id'].apply(lambda x: str(x)+'.png')
test_df['Case ID'] = test_df['Case ID'].apply(lambda x: str(x)+'.png') 

train_df.head()
```

Ahora, limpiaremos la columna de **male**. 
La razón es que la variable booleana no funciona adecuadamente para los propósitos de histogramas, mapeo y conteos.
Se va a sustituir por masculino y femenino.


```python
train_df['gender'] = train_df['male'].apply(lambda x: 'masculino' if x else 'femenino')
print(train_df['gender'].value_counts())

#contamos la cantidad de valores que  hay y hacemos una gráfica
sns.countplot(x = train_df['gender'])
```

# Análisis Exploratorio

## Datos del dataset interesantes

```python
#Paciente mas joven del datasent
print('Edad mínima: ' + str(train_df['boneage'].min()) + ' meses')
```

```python
#Paciente mas grande, en meses
print('Edad máxima: ' + str(train_df['boneage'].max()) + ' meses')
```

```python
#Media de las edades oseas
mediaOseaGeneral = train_df['boneage'].mean()
print('Media osea: ' + str(mediaOseaGeneral))
```

```python
#Desviación estándar
desviacionEdadesOseas = train_df['boneage'].std()
desviacionEdadesOseas
```

Durante una investigación, se encontró que que cuando las features (variables) tienen una desviación estándar pequeña rinden mejor. 
Entonces usaremos Z score para el train
Link: https://sites.google.com/site/geographyfais/fieldwork/6-data-analysis/statistical-tools/prediction

```python
train_df['boneAgeZ'] = (train_df['boneage'] - mediaOseaGeneral)/(desviacionEdadesOseas)
print(train_df.head())
```

Ahora, creamos histogramas para ver como están distribuidos los datos. 
Osea: ver normalidad de los datos y corregir en dado caso sea necesario.

```python
#Graficamos histogramas
train_df['boneage'].hist(color = 'aquamarine')
plt.xlabel('Edad osea en meses')
plt.ylabel('Cantidad de niños')
plt.title('Niños por grupo: masculino o femenino')
```

Vemos que la data muestra una normalidad aceptable. Por tanto, no se deben hacer ajustes en la edad de los niños.

Ahora graficamos el valor Z que ajustamos arriba.

```python
train_df['boneAgeZ'].hist(color = 'orange')
plt.xlabel('Punteo de Edad osea con ajuste Z')
plt.ylabel('Conteo de niños')
plt.title('Relación entre el valor Z y el número de niños.')
```

Arriba recién se graficó el valor de la edad ósea, y notamos normalidad. ¿Pero que hay de la cantidad de datos de niños y niñas?
Los siguientes histogramas muestran la distribución de edad osea **de acorde al género**

```python
male = train_df[train_df['gender'] == 'masculino']
female = train_df[train_df['gender'] == 'femenino']
fig, ax = plt.subplots(2,1)
ax[0].hist(male['boneage'], color = 'aquamarine')
ax[0].set_ylabel('Cantidad de niños')
ax[1].hist(female['boneage'], color = 'violet')
ax[1].set_xlabel('Edad en meses')
ax[1].set_ylabel('Número de niñas')
fig.set_size_inches((10,7))
```

Se nota de nuevo que existe normalidad entre los valores. Vemos que en los varones, la edad osea esta mas acumulada en el lado del 150 meses. Y que en las mujeres se mantiene un poco antes, entre los 140-150 meses aproximadamente.
claro, hay datos atípicos, donde la edad osea es mayor del 200. Y recordemos que el mayor valor es de 228 meses.

Hay un tipo de plot interesante, llamado swarmplot.
Esta dibuja un scatterplot categórico sin puntos que se entrelacen. Da una mejor representación de la distribución de los valores, pero no escala demasiado. 
Es interesante porque muestra todos los datos de forma compacta, como una piscina. 

Extraído de: https://seaborn.pydata.org/generated/seaborn.swarmplot.html

```python
#Relación entre la edad osea y el género
sns.swarmplot(x = train_df['gender'], y = train_df['boneage'])
```

Se nota que la mayoria de los datos de las mujeres estan acumulados en cierta area, y en los varones es mas uniforme. Como siempre esto es para data categórica.


Con fines de que se pueda hacer un análisis de las imágenes, se va a dividir la data en test y valid. Ambas datas parten del train original. El tamaño de la data es del 0.2 para analizar al inicio y que la computación de las imágenes no tarde tanto.

## Haciendo cambios a las imágenes (preprocessing)

Ahora es necesario hacer algunos ajustes a las imágenes.
Esta parte es equivalente a decir: hay que estandarizar las imágenes.
Hay muchas formas. La más sencilla es usar data generators de tensorflow. Lo que hace esto que nos permite hacerle ajustes a las imágenes de acorde a las necesidades. 

Se harán los siguientes ajustes.

* Estandarizar el tamaño de las imágenes a 256x256 pixeles.
* Generar la data de train con columnas de x=id y=boneAgeZ
* Batchsize de 32, es el número de training para cada iteracion.
* Flip_vertical=True, para voltear las imagenes verticalmente.
* Se deja de colores, para eso se coloca el color_mode=rgb
* se ajusta el tamaño.


Extraido de: https://keras.io/api/preprocessing/image/

```python
#importamos librerias
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from  keras.applications.xception import preprocess_input 
```

```python
#Colocamos el tamaño de la imagen que queremos
img_size = 256

train_data_generator = ImageDataGenerator(preprocessing_function = preprocess_input)
val_data_generator = ImageDataGenerator(preprocessing_function = preprocess_input)
```

```python

#Generamos la data de Train con los ajustes que describimos arriba
train_generator = train_data_generator.flow_from_dataframe(
    dataframe = df_train,
    directory = '/kaggle/input/rsna-bone-age/boneage-training-dataset/boneage-training-dataset',
    x_col= 'id',#nombre de la columa x
    y_col= 'boneAgeZ',#Nombre de la columna Y
    batch_size = 32,#batch size
    seed = 25,#semilla de 25 para que sea reproducible
    shuffle = True,#random agarrar
    class_mode= 'other',#el tipo de clase de data, como no es normal, usamos other por ser imagen
    flip_vertical = True,#se hace flip
    color_mode = 'rgb',# las colocamos RGB
    target_size = (img_size, img_size))#colocamos el tamaño de 256x256 pixeles

print("Data de train lograda")
```

Hacemos la misma operación para el test. En caso se quisiera usar en el futuro.


```python
#Data de validación
val_generator = val_data_generator.flow_from_dataframe(
    dataframe = df_valid,
    directory = '/kaggle/input/rsna-bone-age/boneage-training-dataset/boneage-training-dataset',
    x_col = 'id',
    y_col = 'boneAgeZ',
    batch_size = 32,
    seed = 42,
    shuffle = True,
    class_mode = 'other',
    flip_vertical = True,
    color_mode = 'rgb',
    target_size = (img_size, img_size))


#Data de test
test_data_generator = ImageDataGenerator(preprocessing_function = preprocess_input)

test_generator = test_data_generator.flow_from_directory(
    directory = '/kaggle/input/rsna-bone-age/boneage-test-dataset',
    shuffle = True,
    class_mode = None,
    color_mode = 'rgb',
    target_size = (img_size,img_size))
```
